```python
from sklearn.preprocessing import LabelEncoder

# Assume 'df' is the Pandas DataFrame containing the dataset

# Create a LabelEncoder object
le = LabelEncoder()

# Apply the LabelEncoder to the 'Overall Engagement' and 'Job Satisfaction' columns
data1['Overall Engagement'] = le.fit_transform(data1['Overall Engagement'])
data1['Job Satisfaction'] = le.fit_transform(data1['Job Satisfaction'])
```

```python
data1.head()
```

```python
# Get the mapping of labels to codes
engagement_mapping = dict(zip(le.classes_, le.transform(le.classes_)))
print("Overall Engagement Mapping:")
print(engagement_mapping)
# Get the mapping of labels to codes
job_satisfaction_mapping = dict(zip(le.classes_, le.transform(le.classes_)))
print("\nJob Satisfaction Mapping:")
print(job_satisfaction_mapping)
```

```python
overall_engagement_stats = data1["Overall Engagement"].describe()
print("Overall Engagement mean : ",overall_engagement_stats["mean"])
print("Overall Engagement standard deviation : ",overall_engagement_stats["std"])
overall_engagement_mode = data1['Overall Engagement'].mode().iloc[0]
print("Overall Engagement Mode:", overall_engagement_mode)
overall_engagement_median = data1['Overall Engagement'].median()
print("Overall Engagement Median:", overall_engagement_median)
```

```python
job_satisfaction_stats = data1["Job Satisfaction"].describe()
print("Job Satisfaction mean : ",job_satisfaction_stats["mean"])
print("Job Satisfaction standard deviation : ",job_satisfaction_stats["std"])
job_satisfaction_mode = data1['Job Satisfaction'].mode().iloc[0]
print("Job Satisfaction Mode:", job_satisfaction_mode)
job_satisfaction_median = data1['Job Satisfaction'].median()
print("Job Satisfaction Median:", job_satisfaction_median)
```

```python
import matplotlib.pyplot as plt

# Assume 'data1' is the Pandas DataFrame containing the dataset

# Create a scatter plot of Age vs. Job Satisfaction
plt.bar(data1['Age Bracket'], data1['Job Satisfaction'])

# Add title and labels
plt.title('Age vs. Job Satisfaction')
plt.xlabel('Age')
plt.ylabel('Job Satisfaction')

# Show the plot
plt.show()
```
